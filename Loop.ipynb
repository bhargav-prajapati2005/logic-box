2. While loop

```python
i = 1 
while i <= 5:
    print(i)
    i+=1 
```

```python
v = 10
while v > 0:
    print(v)
    v-=1
```

```python
d = 1
while d <= 5:
    print("hello sir")
    d += 1
```

```python
Client = 1000
while Client <= 1000:
    print("big client ")
    Client += 1
```

```python
number = 36
while number <= 40:
    print("good marks ")
    number += 1 
```

```python
i = 1

while i <= 20:
    if i % 2 == 0 and i % 3 == 0:
        print(i, "Divisible by both 2 and 3")
    elif i % 2 == 0:
        print(i, "Divisible by 2")
    elif i % 3 == 0:
        print(i, "Divisible by 3")
    else:
        print(i, "Not divisible by 2 or 3")
    
    i += 1
```

```python
i = 1

while i <= 10:
    if i % 2 == 0:
        i += 1
        continue
    
    print(i)
    i += 1
```

2.Nested Loop

```python
for d in range(10):
    for v in range(1):
        print(d,v)
```

```python
for d in range (1,4):
    for v in range (1,4):
        print(d,v)
```

```python
rows = int(input("enter your rows"))
col = int(input("enter your col"))

for Q in range(rows):
    for K in range(col):
        print("*",end=" ")
    print()
```
